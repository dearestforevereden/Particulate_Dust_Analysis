```python
import pandas as pd

# 파일 경로 설정
csv_21 = '/Users/hayden/Desktop/2021.csv' 
csv_22 = '/Users/hayden/Desktop/2022.csv'
csv_23 = '/Users/hayden/Desktop/2023.csv'

# 각각의 CSV 파일을 데이터프레임으로 읽기
df_2021 = pd.read_csv(csv_21, encoding='cp949')
df_2022 = pd.read_csv(csv_22, encoding='cp949')
df_2023 = pd.read_csv(csv_23, encoding='cp949')

# 데이터프레임을 리스트에 담기
dfss = [df_2021, df_2022, df_2023]

# concat 함수를 사용하여 합치기
weather_df = pd.concat(dfss, ignore_index=True)

# 이름 매핑 사전
name_mapping = {
    '김해시': '김해',
    '양산시': '양산',
    '의령군': '의령',
    '함양군': '함양'
}

weather_df = weather_df.drop(columns=[ '지점'])

# '지점명' 컬럼의 값을 매핑된 값으로 변환
weather_df['지점명'] = weather_df['지점명'].replace(name_mapping)

import pandas as pd
import numpy as np


# 결측치가 있는 열을 선형 보간으로 채우기
for col in weather_df.columns:
    weather_df[col] = weather_df[col].interpolate(method='linear', limit_direction='both')

# # DataFrame 복사본 생성 및 NULL 값 개수 확인
# df_copy = df.copy()

# 각 컬럼의 NULL 값 개수 확인
null_counts = weather_df.isnull().sum()

# 특정 컬럼의 NULL 값 개수 확인
PM25_null_count = weather_df['기온(°C)'].isnull().sum()
PM10_null_count = weather_df['강수량(mm)'].isnull().sum()
O3_null_count = weather_df['풍속(m/s)'].isnull().sum()
CO_null_count = weather_df['습도(%)'].isnull().sum()
SO2_null_count = weather_df['일조(hr)'].isnull().sum()
NO2_null_count = weather_df['전운량(10분위)'].isnull().sum()

print("기온(°C) 컬럼의 NULL 값 개수:", PM25_null_count)
print("강수량(mm) 컬럼의 NULL 값 개수:", PM10_null_count)
print("풍속(m/s) 컬럼의 NULL 값 개수:", O3_null_count)
print("습도(% 컬럼의 NULL 값 개수:", CO_null_count)
print("일조(hr)	컬럼의 NULL 값 개수:", SO2_null_count)
print("전운량(10분위) 컬럼의 NULL 값 개수:", NO2_null_count)

# 컬럼명 매핑 사전
name_mapping = {
    '지점명': 'STTN_NM',
    '일시': 'MSR_DT',
}

# 컬럼명 변경
weather_df.rename(columns=name_mapping, inplace=True)

# 전체 데이터프레임의 NULL 값 개수 확인
total_null_count = weather_df.isnull().sum().sum()
print("전체 데이터프레임의 NULL 값 개수:", total_null_count)

import pandas as pd

# MSR_DT 열을 datetime 형식으로 변환
weather_df['MSR_DT'] = pd.to_datetime(weather_df['MSR_DT'], format='%Y-%m-%d %H:%M')

# 변환된 datetime 형식을 'YYYYMMDDHH' 형식으로 변환
weather_df['MSR_DT'] = weather_df['MSR_DT'].dt.strftime('%Y%m%d%H')

# 변환된 데이터 확인
print(weather_df)


# `MSR_DT` 열을 문자열로 변환
weather_df['MSR_DT'] = weather_df['MSR_DT'].astype(str)

# 두 데이터프레임을 `STTN_NM`과 `MSR_DT`를 기준으로 병합
merged_df = pd.merge(weather_df, df, on=['STTN_NM', 'MSR_DT'], how='inner')

# 병합된 데이터프레임 확인
print(merged_df)


from sklearn.preprocessing import MinMaxScaler
df = merged_df
scaler = MinMaxScaler()
numeric_columns = df.select_dtypes(include=[np.number]).columns

df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

df[numeric_columns]
```

```python
import pandas as pd
import numpy as np
import altair as alt

# MSR_DT를 datetime 형식으로 변환하고 연도, 월, 시간 추출
df['MSR_DT'] = pd.to_datetime(df['MSR_DT'], format='%Y%m%d%H', errors='coerce')
df['Year'] = df['MSR_DT'].dt.year
df['Month'] = df['MSR_DT'].dt.month
df['Hour'] = df['MSR_DT'].dt.hour
df['Hour'] = df['MSR_DT'].dt.hour

# 데이터 필터링 (예시 데이터셋에서는 필터링 없음)
filtered_df = df

# 시간별 평균 계산
yearly_avg = filtered_df.groupby('Hour')[['PM25', 'PM10', 'O3', 'CO', 'SO2', 'NO2']].mean().reset_index()

# 데이터 변환 (melt)
source = yearly_avg.melt(id_vars=['Hour'], 
                         value_vars=['PM25', 'PM10', 'O3', 'CO', 'SO2', 'NO2'],
                         var_name='category', 
                         value_name='value')

# Altair 그래프 생성
base = alt.Chart(source).encode(
    alt.X('Hour:O', title='Hour'),
    alt.Y('value:Q', title='Average Value'),
    alt.Color('category:N', title='Category')
).properties(
    width=800,
    height=400,
    title='Hourly Average of Air Quality Indicators'
)

# 원본 데이터를 사용하여 선 그래프 생성
line = base.mark_line(point=True, size=3)

# 서클 마커 추가
points = base.mark_circle(size=100, opacity=0.7).encode(
    alt.Tooltip(['Hour:N', 'value:Q', 'category:N'])
)

# 그래프 결합
chart = (line + points).resolve_scale(y='independent')  # Y 축 스케일을 독립적으로 설정

chart
```
