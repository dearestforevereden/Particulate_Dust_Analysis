### 1. 기온 (Temperature)

- **의미**: 대기의 온도를 나타내며, 특정 시간과 장소에서의 열 상태를 나타냅니다.
- **단위**: °C (섭씨, Celsius)
    - **의미**: 물의 어는점인 0°C와 끓는점인 100°C를 기준으로 하는 온도 단위입니다.
- **영어 용어**: Temperature

### 2. 강수량 (Precipitation)

- **의미**: 특정 기간 동안 일정한 장소에 내린 비, 눈, 우박 등의 물의 총량을 나타냅니다.
- **단위**: mm (밀리미터, Millimeters)
    - **의미**: 강수량을 수직 깊이로 측정하여, 내린 물의 깊이를 나타내는 단위입니다.
- **영어 용어**: Precipitation

### 3. 풍속 (Wind Speed)

- 
- **의미**: 대기의 움직임 속도, 즉 바람의 속도를 나타냅니다.
- **단위**: m/s (미터/초, Meters per Second)
    - **의미**: 1초 동안 바람이 이동한 거리를 미터 단위로 나타내는 속도 단위입니다.
- **영어 용어**: Wind Speed

### 4. 습도 (Humidity)

- **의미**: 공기 중에 포함된 수증기의 양을 나타내며, 공기의 습윤 정도를 나타냅니다.
- **단위**: % (퍼센트, Percent)
    - **의미**: 포화 수증기량에 대한 실제 수증기량의 비율을 백분율로 나타낸 것입니다.
- **영어 용어**: Humidity

### 5. 일조 (Sunshine Duration)

- **의미**: 태양이 비추는 시간으로, 일사량과 관련된 정보를 제공하여, 햇빛이 비치는 총 시간을 나타냅니다.
- **단위**: hr (시간, Hours)
    - **의미**: 하루 동안 햇빛이 비친 총 시간을 시간 단위로 측정합니다.
- **영어 용어**: Sunshine Duration

### 6. 전운량 (Total Cloud Cover)

- **의미**: 하늘에 구름이 덮여 있는 정도를 나타내며, 하늘 전체를 10분위로 나눈 값입니다.
- **단위**: 10분위 (Tenths)
    - **의미**: 하늘 전체를 10등분하여, 구름이 덮인 부분의 비율을 나타냅니다. 0은 맑음, 10은 완전히 흐림을 의미합니다.
- **영어 용어**: Total Cloud Cover

```python
from sklearn.model_selection import train_test_split

# 특성과 타겟 변수 정의
X = df[['기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)',
        '일조(hr)', '적설(cm)', '전운량(10분위)', 'PM25',	'PM10',	'O3',	'CO',	'SO2',	'NO2']]
y_pm25 = df['PM25']
y_pm10 = df['PM10']

# 훈련 세트와 테스트 세트로 분리
X_train, X_test, y_train_pm25, y_test_pm25 = train_test_split(X, y_pm25, test_size=0.2, random_state=42)
X_train, X_test, y_train_pm10, y_test_pm10 = train_test_split(X, y_pm10, test_size=0.2, random_state=42)
```

LinearRegression: 선형 회귀 모델을 사용하여 PM25와 PM10을 예측합니다. 선형 회귀는 특성과 타겟 변수 간의 선형 관계를 학습합니다.
fit: 훈련 세트(X_train, y_train_pm25 또는 y_train_pm10)를 사용하여 모델을 학습합니다.
predict: 테스트 세트(X_test)에 대해 예측을 수행합니다. 이 예측 결과를 y_pred_pm25와 y_pred_pm10에 저장합니다.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# PM25 예측 모델
model_pm25 = LinearRegression()
model_pm25.fit(X_train, y_train_pm25)
y_pred_pm25 = model_pm25.predict(X_test)

# PM10 예측 모델
model_pm10 = LinearRegression()
model_pm10.fit(X_train, y_train_pm10)
y_pred_pm10 = model_pm10.predict(X_test)

# 성능 검증
print("PM25 Model Performance:")
print(f"Mean Squared Error: {mean_squared_error(y_test_pm25, y_pred_pm25)}")
print(f"R^2 Score: {r2_score(y_test_pm25, y_pred_pm25)}")

print("PM10 Model Performance:")
print(f"Mean Squared Error: {mean_squared_error(y_test_pm10, y_pred_pm10)}")
print(f"R^2 Score: {r2_score(y_test_pm10, y_pred_pm10)}")


```

Mean Squared Error (MSE): 평균 제곱 오차는 예측값과 실제값 간의 차이를 제곱하여 평균한 값입니다. 값이 작을수록 모델의 성능이 좋다는 것을 의미합니다. MSE는 예측 오차의 제곱 평균을 측정합니다.

R^2 Score: 결정 계수는 모델이 종속 변수의 분산을 얼마나 잘 설명하는지를 나타내는 지표입니다. 값의 범위는 0과 1 사이입니다. 1에 가까울수록 모델이 데이터를 잘 설명한다는 것을 의미합니다.

모델의 성능이 매우 우수하게 나왔지만, 오버피팅과 데이터의 일반화 능력을 고려하여 추가적인 검토(교차 검증(Cross-validation): 교차 검증을 통해 데이터의 다른 부분에서도 모델이 잘 작동하는지 확인하여 오버피팅 여부를 검토할 수 있습니다.)가 필요할 수 있습니다.

교차 분석과 산점도 그래프를 통해 PM25와 PM10 사이의 관계를 이해하는 것은 데이터 분석에서 중요한 부분입니다. 이 그래프는 두 변수 사이의 상관관계를 시각적으로 표현합니다. 아래는 이 그래프를 해석하고 좋은 분석 결과를 얻기 위한 방법입니다.

산점도 그래프 설명
산점도 그래프의 기본 설명:

X축: PM25의 실제값
Y축: PM10의 실제값
점의 위치: 각 점은 특정 시간과 지역에 대한 PM25와 PM10의 실제값을 나타냅니다.
점의 투명도 (alpha): 점이 너무 밀집될 경우, alpha 값을 조정하여 시각적으로 밀도를 줄일 수 있습니다.
해석 방법:

상관관계 (Correlation):

양의 상관관계 (Positive Correlation): PM25와 PM10이 함께 증가하거나 감소하는 경향이 보이면, 두 변수 사이에 양의 상관관계가 있다는 의미입니다. 그래프에서 점들이 대각선 방향으로 배열되는 경향이 보이면, 양의 상관관계가 있을 가능성이 높습니다.
음의 상관관계 (Negative Correlation): PM25가 증가할 때 PM10이 감소하는 경향이 보이면, 음의 상관관계가 있다는 의미입니다. 그래프에서 점들이 반대 방향으로 배열되는 경향이 보이면, 음의 상관관계가 있을 수 있습니다.
상관관계 없음 (No Correlation): 점들이 무작위로 퍼져 있거나 특정 패턴이 없으면, 두 변수 사이에 상관관계가 없다는 의미입니다.
선형성 (Linearity):

점들이 선형적인 패턴을 보이면, 두 변수 사이에 선형적인 관계가 있을 수 있습니다.
비선형적인 패턴이 보이면, 두 변수 사이의 관계가 비선형적일 수 있습니다.
클러스터 (Clusters):

점들이 특정 클러스터를 형성하면, 이는 특정 시간대나 지역에서 PM25와 PM10의 값이 비슷하게 나타날 수 있음을 의미할 수 있습니다.
이상치 (Outliers):

그래프에서 다른 점들과 떨어져 있는 점들이 있다면, 이는 이상치로 간주될 수 있습니다. 이상치는 데이터 분석에서 주의가 필요할 수 있습니다.
좋은 그래프의 특징
명확한 패턴: 두 변수 사이에 명확한 패턴이 보인다면, 이는 상관관계가 있음을 시사합니다. 예를 들어, 점들이 대각선 방향으로 배열된다면 양의 상관관계가 있음을 나타냅니다.

균일한 분포: 데이터가 균일하게 분포되어 있다면, 두 변수 간의 관계가 잘 표현된 것입니다. 특정 영역에 데이터가 집중되거나 빈약하게 나타나면 추가 분석이 필요할 수 있습니다.

이상치 식별: 이상치가 명확히 식별되면, 이들을 분석하여 이상치가 데이터의 특성인지, 아니면 데이터 입력 오류인지 확인할 수 있습니다.

교차 분석과 상관관계 분석
교차 분석: PM25와 PM10의 관계를 분석하는 과정으로, 두 변수의 관계를 이해하기 위한 첫 단계입니다. 이 그래프는 두 변수 간의 관계를 직관적으로 보여줍니다.

상관계수 (Correlation Coefficient): 산점도 그래프와 함께 상관계수를 계산하여 두 변수 사이의 관계를 수치적으로 파악할 수 있습니다. 상관계수는 -1과 1 사이의 값을 가지며, 1에 가까울수록 강한 양의 상관관계, -1에 가까울수록 강한 음의 상관관계를 나타냅니다.

```python
# 시간 변수를 숫자로 변환 (Unix time)
df['MSR_DT'] = pd.to_datetime(df['MSR_DT'], format='%Y%m%d%H')
df['Time'] = (df['MSR_DT'].astype(int) / 10**9).astype(int)  # Unix time

# 특성과 타겟 변수 정의
X = df[['기온(°C)', '강수량(mm)', '풍속(m/s)', '습도(%)',
        '일조(hr)', '적설(cm)', '전운량(10분위)', 'O3', 'CO', 'SO2', 'NO2']]
y_pm25 = df['PM25']
y_pm10 = df['PM10']

# 훈련 세트와 테스트 세트로 분리
X_train, X_test, y_train_pm25, y_test_pm25 = train_test_split(X, y_pm25, test_size=0.2, random_state=42)
X_train, X_test, y_train_pm10, y_test_pm10 = train_test_split(X, y_pm10, test_size=0.2, random_state=42)

# PM25 예측 모델
model_pm25 = LinearRegression()
model_pm25.fit(X_train, y_train_pm25)

# PM10 예측 모델
model_pm10 = LinearRegression()
model_pm10.fit(X_train, y_train_pm10)

# 미래 데이터 생성 (6개월간의 시간 데이터 생성)
future_times = pd.date_range(start='2024-01-01', periods=180*24, freq='H')  # 6개월간의 시간 데이터 생성
future_times_unix = (future_times.astype(int) / 10**9).astype(int)  # Unix time

# 과거 데이터에서 평균값으로 특성 데이터 생성
average_values = {
    '기온(°C)': df['기온(°C)'].mean(),
    '강수량(mm)': df['강수량(mm)'].mean(),
    '풍속(m/s)': df['풍속(m/s)'].mean(),
    '습도(%)': df['습도(%)'].mean(),
    '일조(hr)': df['일조(hr)'].mean(),
    '적설(cm)': df['적설(cm)'].mean(),
    '전운량(10분위)': df['전운량(10분위)'].mean(),
    'O3': df['O3'].mean(),
    'CO': df['CO'].mean(),
    'SO2': df['SO2'].mean(),
    'NO2': df['NO2'].mean()
}

# 미래 데이터에 대해 평균값을 입력으로 사용
future_X = pd.DataFrame({
    '기온(°C)': [average_values['기온(°C)']] * len(future_times),
    '강수량(mm)': [average_values['강수량(mm)']] * len(future_times),
    '풍속(m/s)': [average_values['풍속(m/s)']] * len(future_times),
    '습도(%)': [average_values['습도(%)']] * len(future_times),
    '일조(hr)': [average_values['일조(hr)']] * len(future_times),
    '적설(cm)': [average_values['적설(cm)']] * len(future_times),
    '전운량(10분위)': [average_values['전운량(10분위)']] * len(future_times),
    'O3': [average_values['O3']] * len(future_times),
    'CO': [average_values['CO']] * len(future_times),
    'SO2': [average_values['SO2']] * len(future_times),
    'NO2': [average_values['NO2']] * len(future_times)
})

# 예측 수행
future_pm25 = model_pm25.predict(future_X)
future_pm10 = model_pm10.predict(future_X)

# 결과 시각화
plt.figure(figsize=(14, 7))

# PM25 예측 결과 시각화
plt.subplot(1, 2, 1)
plt.plot(future_times, future_pm25, label='Predicted PM25', color='blue')
plt.xlabel('Date')
plt.ylabel('PM25')
plt.title('PM25 Forecast')
plt.legend()

# PM10 예측 결과 시각화
plt.subplot(1, 2, 2)
plt.plot(future_times, future_pm10, label='Predicted PM10', color='red')
plt.xlabel('Date')
plt.ylabel('PM10')
plt.title('PM10 Forecast')
plt.legend()

plt.tight_layout()
plt.show()
```
